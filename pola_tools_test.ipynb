```python
# testing with the whole cube
import pickle

with open("example_cube_cards.pkl", 'rb') as f:
    cards = pickle.load(f)
trainer_cards = [card for card in cards if card.supertype == "Trainer"]
trainer_cards[0].name
```

```python
# trying to classify all trainers
tool = PolaTool(gemini_model='gemini-2.0-flash-lite')
results = tool.classify_trainers(trainer_cards)
```

```python
for i, result in enumerate(results):
    print(i, result)
```
